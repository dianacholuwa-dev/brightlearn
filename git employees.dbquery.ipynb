```sql
---QUERY 1:RETRIEVE ALL RECORDS
Select *
 from employees
  Limit 10;

  ---QUERY 2:distinct departments
  SELECT DISTINCT department
  from employees;

  ---QUERY 3:order by salary
  SELECT First_name,LAST_NAME,SALARY
  FROM employees 
  order by salary desc;

  ---QUERY 4:limit startement
  SELECT id,First_name,LAST_NAME,SALARY
  FROM employees 
  order by salary desc
  limit 5;



  ---QUERY 5:WHERE STATEMENT
  SELECT *
  FROM employees
  where Department ='IT';

---Query 6:AND STATEMENT
SELECT *
FROM employees
WHERE Department ='Finance'AND SALARY>58000;

---Query 7:OR STATEMET
SELECT *
FROM employees
WHERE Department ='HR'OR Department = 'Marketing';

---Qeury 8:NOT Statement
select *
From employees
where NOT Department ='IT';

---QUERY 9:IN Statement
select *
From employees 
where Department IN ('HR','IT','Finance');

---QUERY 10:COMBINING STATEMENT
SELECT *
FROM employees
WHERE city ='New York' AND Department ='IT' AND SALARY>50000;

---QUERY 11:COMBING WHERE,AND,and ORDER BY
SELECT First_Name,LAST_NAME,SALARY
FROM employees
WHERE Department ='Finance' OR Department='Marketing'
ORDER BY SALARY>52000 DESC

---QUERY 12:COMBING SELECT DISTINCT,WHERE and IN
Select DISTINCT City
From employees
WHERE Department NOT in('HR') and Department NOT in('IT');






 
```
